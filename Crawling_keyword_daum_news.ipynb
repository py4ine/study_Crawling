```python
import requests
from bs4 import BeautifulSoup
import re
from konlpy.tag import Okt # 형태소 분석기
from collections import Counter
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
```

```python
# 다음뉴스 크롤링 함수
titleList = []

def cr_title(keyword, page):

    for i in range(1, page+1, 1):
        url = f"https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q={keyword}&p={i}&sort=recency"
        print(url)
        response = requests.get(url)
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')

        titles = soup.select('ul.c-list-basic')  # 선택할 ul 태그 지정

        for title in titles:
            titleText = title.text  # 타이틀의 text만 추출
            print(titleText)
            titleList.append(titleText)  #추출한 타이틀text를 titleList에 추가

        time.sleep(2)
```

```python
# 크롤링 (keyword, 페이지수)
cr_title("유통", 50)
```

```python
# 크롤링 텍스트 내용 확인
titleList
```

```python
# (정규표현식 활용하여) 한글만 추출
def text_cleaning(text):
    korean = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = korean.sub('', text)
    return result
```

```python
# 전처리: 한글만 추출
# 리스트로 바꿔줄 / map: 여러개의 데이터를, / lambda: 함수정의 대신
titleList = list(map(lambda x:text_cleaning(x), titleList))
```

```python
# 전처리: 명사 빈도수 계산
title_corpus = " ".join(titleList)  # 리스트를 문자열로 변환
nouns_tagger = Okt()  # 문자열을 명사 단위로 추출하는 함수
nouns = nouns_tagger.nouns(title_corpus)  # 명사추출 함수 사용해 분류
count_nouns = Counter(nouns)  # 각 명사의 빈도수 계산
```

```python
# 추출한 명사들의 빈도수 확인
print(count_nouns)
print(len(count_nouns))
```

```python
# 전처리: 불용어 처리
remove_char_counter = Counter({x:count_nouns[x] for x in count_nouns if len(x) > 1})
print(len(remove_char_counter))
```

```python
korean_stopwords_path = "data/stopwords-ko.txt"  # 불용어 텍스트 파일 경로

# 불용어 로드
with open(korean_stopwords_path, encoding='utf8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

# 나무위키에서 추가한 불용어
namu_wiki_stopwords = ['상위', '문서', '내용', '누설', '아래', '해당', '설명',
                       '표기', '추가', '모든', '사용', '매우', '가장', '줄거리',
                       '요소', '상황', '편집', '틀', '경우', '때문', '모습',
                       '정도', '이후', '사실', '생각', '인물', '이름', '년월', '공유', 
                       '인공', '지능'] ##### 개별적 불용어 추가

# 추가 불용어 설정
self_add_stopwords = ['메뉴', '개별', '간전', '언론사', '기자'

]

# 불용어 추가
stopwords.extend(namu_wiki_stopwords)
stopwords.extend(self_add_stopwords)

# 불용어 처리
remove_char_counter = Counter({x:remove_char_counter[x] for x in remove_char_counter if x not in stopwords})
```

```python
# 텍스트 전처리: 명사의 분석 대상 갯수 지정
ranked_tags = remove_char_counter.most_common(50)  # 빈도순 n개 단어
pd.DataFrame(ranked_tags, columns=('keywords', 'qty'))
```
